### **CNN - Calculating the Number of parameters:**

**1. Convolutional Neurons-Cont**:

A CNN have an input volume of size [16x16x20]. If the receptive field of the filter in the ensuing convolutional layer is 3x3, what would be the number of its connections (weights)?
IMG_2329.jpg

  **Solution:**
  - Input volume: [16x16x20]
  - Receptive field (filter size): [3x3] scanning a region [3x3x20] 
  - Each filter has a bias parameter 
  → Single filter weight = ([3x3]+1)x20 = 200


**2. CNN - Output volume size**
A CNN layer has a 7x7 input feeding ten 3x3 Conv filters with stride 2 and zero pad 1. What would be the size of the output volume after a 2x2 max pool layer with no overlap?  
Note: a 2x2 max pool layer with no overlapmeans that we divide the input image into non-overlapping 2x2 regions and take the maximum value within each region as the output value for that region

**Solution:**

Summary CNN parameter calculation:
- Accepts a volume of size $W_1\times H_1 \times D_1$ = Input [7x7x10] 
- Require four hyperprameters:
  - Number of filters $K$
  - Spatial extent $F \ (\text{receptive field size}) = 3$
  - The stride $S=2$
  - The amount of zero padding $P =1$ 
- Produces of a volume size $W_2 \times H_2 \times D_2$ where:
  - $W_2 = \frac{W_1-F+2P}{S}+1 = \frac{7-3+2.(1)}{2} +1 = 4$
  - $H_2 = \frac{H_1 -F +2P}{S}+1 = 4$
  - $D_2 = K $

→ Output of Conv layer $=[4 \times 4 \times 10]$ 
- Applying a $2\times 2$ max pool layer with no overlap

→ Output size $=\frac{4-2}{2}+1=2$
Volume of the max pool layer is $=[2 \times 2 \times 10]$
